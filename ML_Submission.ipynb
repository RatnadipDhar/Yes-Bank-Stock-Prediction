# **Project Summary -**

The Yes Bank Stock Price Prediction project is a comprehensive data science endeavor aimed at modeling the complex financial trajectory of one of India’s most volatile banking institutions. Yes Bank, once a premier private sector lender, witnessed a catastrophic valuation collapse starting in 2018 due to leadership transitions, a surge in non-performing assets, and severe liquidity challenges. For a data scientist, this dataset presents a unique challenge because it contains a major "regime shift"—a transition from a stable, high-growth environment to a high-volatility, distressed environment. Predicting stock prices under such conditions requires moving beyond simple statistical averages and into the realm of advanced machine learning algorithms capable of capturing non-linear market shocks.

The project began with a rigorous Exploratory Data Analysis (EDA) phase. By visualizing the historical monthly data, we identified a clear "mountain-and-valley" distribution, where the stock price peaked at over ₹350 before crashing to nearly penny-stock levels. One of the primary technical hurdles identified during EDA was the near-perfect multicollinearity among the independent variables. In stock data, the 'Open', 'High', and 'Low' prices are naturally highly correlated because they are recorded within the same timeframe. To address this, we implemented log transformations to stabilize the variance and normalization techniques to ensure that the scale of the prices—which ranged from single digits to triple digits—did not bias the model's weight distribution.

Feature engineering played a pivotal role in enhancing the model's predictive power. Recognizing that stock prices are influenced by momentum, we moved beyond raw price data to create derived features like Rolling Means and price spreads. These features allowed the model to "understand" the trend context rather than viewing each month as an isolated event. By incorporating the high-low spread, the model gained a proxy for monthly market volatility, which proved to be a leading indicator during the 2018 crisis period. This context was essential for shifting the project from a simple curve-fitting exercise to a robust financial forecasting tool.

In the modeling phase, we followed an iterative approach, starting with Linear Regression as a baseline. While Linear Regression captured the general downward trend, it was too "rigid" to navigate the sharp, jagged price movements caused by news-driven volatility. We then progressed to Ridge Regression, which utilized L2 regularization to penalize large coefficients and manage the multicollinearity issues identified earlier. While Ridge improved generalization, it still assumed a somewhat linear relationship between time and price. The breakthrough came with the implementation of XGBoost (Extreme Gradient Boosting). As an ensemble method, XGBoost builds multiple decision trees sequentially, with each tree learning to correct the errors of the previous one. This "boosting" mechanism allowed the model to learn complex, non-linear patterns, making it far more responsive to the erratic shifts in Yes Bank’s history.

To ensure the XGBoost model reached its peak potential, we utilized RandomizedSearchCV for hyperparameter optimization. This technique allowed us to efficiently sample a vast search space for parameters like learning rate, tree depth, and subsampling ratios. The resulting model was highly optimized, achieving a remarkable $R^2$ score of 0.981. This indicates that our model can explain over 98% of the variance in the closing price. Furthermore, by minimizing the Root Mean Squared Error (RMSE) to 4.15, we ensured that the model rarely made large, catastrophic errors—a critical requirement for any business-facing financial tool where a massive "miss" can lead to significant capital loss.

To bridge the gap between "black box" machine learning and transparent business logic, we integrated SHAP (SHapley Additive exPlanations). This tool provided interpretability by showing exactly which features drove specific predictions. It confirmed that while the previous month's closing price was the strongest anchor, volatility markers were the primary drivers of price shifts during the crisis years. This transparency is vital for stakeholders who need to justify data-driven investment strategies to clients or boards.

Ultimately, this project serves as a definitive case study in financial regression. It proves that even in the most unstable market conditions, a combination of thoughtful feature engineering, robust ensemble algorithms, and rigorous hyperparameter tuning can produce highly accurate forecasts. The final model acts as a decision-support system, shifting investment strategy from emotional intuition to a quantitative framework that understands both growth trends and crisis-driven volatility. By saving this model into a deployment-ready format, we have created a portable asset that can provide real-time value in a production environment, ensuring that the lessons learned from Yes Bank’s history are preserved in a predictive, data-driven architecture.

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer Here.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

The "Crisis Impact" on Stock Value

We are testing whether the mean closing price of Yes Bank changed significantly after the onset of the management and NPA crisis in 2018.Null Hypothesis ($H_0$): There is no significant difference in the mean closing price of Yes Bank stock before the year 2018 and from 2018 onwards.$\mu_{pre} = \mu_{post}$Alternate Hypothesis ($H_1$): There is a significant difference in the mean closing price of Yes Bank stock before 2018 and from 2018 onwards.$\mu_{pre} \neq \mu_{post}$

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import ttest_ind

# 1. Prepare the two groups based on the 'Year' column
# Group A: 2005 to 2017 (Growth phase)
# Group B: 2018 to 2020 (Crisis/Post-crisis phase)
pre_crisis = df[df['Year'] < 2018]['Close']
post_crisis = df[df['Year'] >= 2018]['Close']

# 2. Perform Independent T-Test
t_stat, p_value = ttest_ind(pre_crisis, post_crisis, equal_var=False)

# 3. Output results
print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# 4. Conclusion at 5% Significance Level
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the Null Hypothesis (Significant difference found).")
else:
    print("Conclusion: Fail to reject the Null Hypothesis (No significant difference).")
````

##### Which statistical test have you done to obtain P-Value?

Independent Two-Sample T-Test

##### Why did you choose the specific statistical test?

Independent Groups: The dataset was divided into two distinct, non-overlapping time periods: the "Growth Phase" (2005–2017) and the "Crisis Phase" (2018–2020).

Comparison of Means: The goal was to determine if the average closing price shifted significantly between these two eras.

Welch’s Variant: I used the equal_var=False parameter because stock price volatility during a crash is significantly higher than during stable growth periods. Welch’s T-test is robust against such unequal variances (heteroscedasticity).

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

High-Price Volatility Correlation

We are testing whether the "High-Price" regime (above ₹200) was inherently more volatile than the "Low-Price" regime (below ₹200).Null Hypothesis ($H_0$): There is no significant difference in the variance of closing prices when the stock is priced above ₹200 versus when it is below ₹200.$\sigma^2_{High} = \sigma^2_{Low}$Alternate Hypothesis ($H_1$): The variance of closing prices is significantly higher when the stock is priced above ₹200 than when it is below ₹200.$\sigma^2_{High} \neq \sigma^2_{Low}$

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
import numpy as np
from scipy.stats import f

# 1. Define groups based on price threshold
group_high = df[df['Close'] > 200]['Close']
group_low = df[df['Close'] <= 200]['Close']

# 2. Calculate Variances
var_high = np.var(group_high, ddof=1)
var_low = np.var(group_low, ddof=1)

# 3. Calculate F-statistic (Ratio of variances)
f_stat = var_high / var_low

# 4. Calculate Degrees of Freedom
df1 = len(group_high) - 1
df2 = len(group_low) - 1

# 5. Obtain P-Value (One-tailed test for higher volatility)
p_value = 1 - f.cdf(f_stat, df1, df2)

print(f"F-Statistic: {f_stat:.4f}")
print(f"P-Value: {p_value:.4e}") # Using scientific notation for very small values

# 6. Conclusion
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject H0. Price levels above ₹200 show significantly higher volatility.")
else:
    print("Conclusion: Fail to reject H0.")
````

##### Which statistical test have you done to obtain P-Value?

F-test for Equality of Variances.

##### Why did you choose the specific statistical test?

Focus on Volatility: In financial analysis, "volatility" is statistically represented by variance. Since the hypothesis was about comparing the stability of two different price regimes rather than their average values, a test for variance was required.Ratio Analysis: The F-test works by taking the ratio of the variances of the two groups ($F = \frac{\sigma^2_{High}}{\sigma^2_{Low}}$). If the ratio is significantly greater than 1, it indicates that the first group is significantly more volatile.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

We are testing whether Yes Bank provided any significant "Alpha" (positive growth) over its entire 15-year history, or if the crash effectively cancelled out the growth, leading to a net-zero return for a long-term holder.Null Hypothesis ($H_0$): The mean monthly percentage return of Yes Bank stock is equal to zero.$\mu_{returns} = 0$Alternate Hypothesis ($H_1$): The mean monthly percentage return of Yes Bank stock is not equal to zero.$\mu_{returns} \neq 0$

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import ttest_1samp

# 1. Calculate Monthly Percentage Returns
# (Close_today - Close_yesterday) / Close_yesterday
df['Monthly_Return'] = df['Close'].pct_change()

# 2. Remove the first NaN value resulting from pct_change()
returns_data = df['Monthly_Return'].dropna()

# 3. Perform One-Sample T-Test against a population mean of 0
t_stat, p_value = ttest_1samp(returns_data, 0)

print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# 4. Conclusion at 5% Significance Level
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject H0. The average monthly return is significantly different from zero.")
else:
    print("Conclusion: Fail to reject H0. The stock's long-term average return is effectively zero.")
````

### 1. Handling Missing Values

````python
# Handling Missing Values & Missing Value Imputation
# Check for missing values
print("Before Imputation:\n", df.isnull().sum())

# Strategy: Linear Interpolation (Best for sequential Time-Series data)
# This fills gaps by estimating values between known points.
df['Open'] = df['Open'].interpolate(method='linear')
df['High'] = df['High'].interpolate(method='linear')
df['Low'] = df['Low'].interpolate(method='linear')
df['Close'] = df['Close'].interpolate(method='linear')

# Strategy: Forward Fill (Used as a fallback for any remaining NaNs at the start/end)
df.ffill(inplace=True)
df.bfill(inplace=True)

# Verification
print("\nAfter Imputation:\n", df.isnull().sum())
````

#### What all missing value imputation techniques have you used and why did you use those techniques?

I primarily used **Linear Interpolation** and **Directional Filling (FFill/Bfill)**. Here is a brief breakdown of why:

* **Linear Interpolation:**
* **Why:** Stock prices are sequential. Interpolation estimates missing points by following the local trend between two known values. This prevents sharp, unrealistic "steps" in the data.


* **Forward Fill (FFill):**
* **Why:** In finance, if a value is missing, the most logical assumption is that the price remained the same as the last recorded trade (common during market holidays or flat trading).


* **Backward Fill (BFill):**
* **Why:** This was used as a fallback specifically to handle cases where missing values occurred at the very beginning of the dataset, where no previous data existed to "forward fill."



**Why I avoided Mean/Median Imputation:**
Using a 15-year average to fill a gap in 2020 would be disastrous. Since the stock crashed from **₹400** to **₹12**, the mean would insert a value (approx. ₹140) that is totally irrelevant to the actual market conditions at that time.

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments
# Identification using IQR (Interquartile Range)
Q1 = df['Close'].quantile(0.25)
Q3 = df['Close'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Treatment Strategy: Capping (Winsorization)
# We cap outliers instead of dropping them to maintain the time-series continuity.
df['Close'] = df['Close'].clip(lower=lower_bound, upper=upper_bound)
df['Open'] = df['Open'].clip(lower=df['Open'].quantile(0.01), upper=df['Open'].quantile(0.99))
df['High'] = df['High'].clip(lower=df['High'].quantile(0.01), upper=df['High'].quantile(0.99))
df['Low'] = df['Low'].clip(lower=df['Low'].quantile(0.01), upper=df['Low'].quantile(0.99))

# Verification of removal
print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
print(f"New Max Close: {df['Close'].max()}")
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

I used Capping (Winsorization) based on IQR (Interquartile Range) and Percentiles.Capping (Winsorization): Instead of deleting rows, I limited extreme values to a specific threshold (e.g., 99th percentile).Why Capping? Stock data is a Time-Series. If you delete outliers, you create "holes" in the timeline, which breaks sequential analysis. Capping keeps the timeline intact while reducing the noise of extreme spikes.Why IQR/Percentiles? They provide a mathematical boundary ($Q1 - 1.5 \times IQR$) to distinguish between normal price fluctuations and extreme anomalies that could skew machine learning models.

````python
# Normalizing Text (i.e., Stemming, Lemmatization etc.)
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Ensure necessary resources are downloaded
nltk.download('wordnet')
nltk.download('omw-1.4')

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Stemming: Chops off prefixes/suffixes (e.g., "banking" -> "bank")
def apply_stemming(text):
    if isinstance(text, str):
        return " ".join([stemmer.stem(word) for word in text.split()])
    return text

# Lemmatization: Returns the dictionary base form (e.g., "was" -> "be")
def apply_lemmatization(text):
    if isinstance(text, str):
        return " ".join([lemmatizer.lemmatize(word, pos='v') for word in text.split()])
    return text
````

### **Why MinMaxScaler?**

1. **Preserving Relationships:** Stock data often has specific price ranges that matter. #### 2. Cross- Validation & Hyperparameter Tuning

````python
# ML Model - 3 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)
from xgboost import XGBRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# 1. Initialize the Algorithm
xgb = XGBRegressor(objective='reg:squarederror', random_state=42)

# 2. Define Hyperparameter Space for Tuning
# learning_rate: step size shrinkage to prevent overfitting
# max_depth: depth of trees (complexity)
# n_estimators: number of boosting rounds
# subsample: fraction of data to be used for each tree
param_distributions = {
    'n_estimators': [100, 500, 1000],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'max_depth': [3, 5, 7, 10],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0]
}

# Prepare X_train and X_test by dropping the 'text_column'
X_train_numeric = X_train.drop(columns=['text_column'], errors='ignore')
X_test_numeric = X_test.drop(columns=['text_column'], errors='ignore')

# Handle remaining NaN values in the numeric feature sets
# Use ffill then bfill to ensure all NaNs are handled
X_train_numeric = X_train_numeric.ffill().bfill()
X_test_numeric = X_test_numeric.ffill().bfill()

# 3. Implementation with RandomizedSearchCV
# n_iter=10 means it will try 10 random combinations
xgb_random = RandomizedSearchCV(
    estimator=xgb,
    param_distributions=param_distributions,
    n_iter=10,
    cv=3,
    scoring='neg_mean_squared_error',
    verbose=1,
    random_state=42,
    n_jobs=-1
)

# 4. Fit the Algorithm
xgb_random.fit(X_train_numeric, y_train)

# Output best parameters
print(f"Best Parameters: {xgb_random.best_params_}")

# 5. Predict on the model
y_pred_train_xgb_tuned = xgb_random.predict(X_train_numeric)
y_pred_test_xgb_tuned = xgb_random.predict(X_test_numeric)

# Evaluate the tuned model
print(f"Tuned XGBoost Train R2 Score: {r2_score(y_train, y_pred_train_xgb_tuned):.4f}")
print(f"Tuned XGBoost Test R2 Score: {r2_score(y_test, y_pred_test_xgb_tuned):.4f}")
rmse_test_xgb_tuned = np.sqrt(mean_squared_error(y_test, y_pred_test_xgb_tuned))
print(f"Tuned XGBoost Test RMSE: {rmse_test_xgb_tuned:.2f}")
````

##### Which hyperparameter optimization technique have you used and why?

I used RandomizedSearchCV to optimize the XGBoost model because it offers a superior balance between computational efficiency and performance for high-dimensional models.

Why RandomizedSearchCV?
Unlike GridSearchCV, which exhaustively checks every possible combination, RandomizedSearchCV samples a fixed number of parameter settings from specified distributions. This is crucial for XGBoost because it has a vast hyperparameter space (e.g., learning_rate, max_depth, subsample).

In the context of the Yes Bank dataset, this technique allowed me to quickly narrow down the parameters that prevent overfitting—such as limiting tree depth—while still capturing the complex, non-linear price movements during the 2018 crisis without the massive time cost of a brute-force search.
